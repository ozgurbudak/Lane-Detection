```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
import numpy as np
import imageio
import os
from os.path import isfile, join
```

```python

image_list= []
reader = imageio.get_reader('solidYellowLeft.mp4')
count=0
for i, im in enumerate(reader):
    print('Extracted frame %i ' % (i))
    name = str(count) + '.jpg'
    image_list.append(name)
    imageio.imwrite(name, im)
    count +=1
```

```python
def findLanesInImage(s):
    image = mpimg.imread(s)
    
    gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY) #grayscale conversion
    
    kernel_size = 5
    blur_gray = cv2.GaussianBlur(gray,(kernel_size, kernel_size), 0)
    
    low_threshold = 150
    high_threshold = 300
    edges = cv2.Canny(blur_gray, low_threshold, high_threshold)
    
    mask = np.zeros_like(edges)   
    ignore_mask_color = 255 
    
    imshape = image.shape
    vertices = np.array([[(0,imshape[0]),(450, 290), (490, 290), (imshape[1],imshape[0])]], dtype=np.int32)
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    masked_edges = cv2.bitwise_and(edges, mask)
    
    rho = 2 # distance resolution in pixels of the Hough grid
    theta = np.pi/180 # angular resolution in radians of the Hough grid
    threshold = 5     # minimum number of votes (intersections in Hough grid cell)
    min_line_length = 100 #minimum number of pixels making up a line
    max_line_gap = 100  # maximum gap in pixels between connectable line segments
    line_image = np.copy(image)*0 # creating a blank to draw lines on

    lines = cv2.HoughLinesP(masked_edges, rho, theta, threshold, np.array([]),
                            min_line_length, max_line_gap)
    
    
    if lines is None:
        return image
    else:
        for line in lines:
            for x1,y1,x2,y2 in line:
                cv2.line(line_image,(x1,y1),(x2,y2),(255,0,0),10)
      
        color_edges = np.dstack((edges, edges, edges))     
        lines_edges = cv2.addWeighted(image, 0.8, line_image, 1, 0)
        return lines_edges
    
```

```python
image_arr=[]

for name in image_list:
    print('Finding lane in', name)
    x= findLanesInImage(name)
    image_arr.append(x)
    height, width, layers = x.shape
    size = (width,height)
    imageio.imwrite(name, x)
```

```python
out = cv2.VideoWriter('answer.mp4',cv2.VideoWriter_fourcc(*'X264'), 30, size)
for i in range(len(image_arr)):
    # writing to a image array
    out.write(image_arr[i])
out.release()
```
